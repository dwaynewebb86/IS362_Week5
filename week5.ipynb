Question 1: What is the northernmost airport in the United States?

Question 2: What is the easternmost airport in the United States?


```python
import pandas as pd

# Load CSV
df = pd.read_csv("airports.csv")

# Northernmost airport
northernmost = df.loc[df["lat"].idxmax()]

# Easternmost airport
easternmost = df.loc[df["lon"].idxmax()]

print("Northernmost Airport:")
print(northernmost)

print("\nEasternmost Airport:")
print(easternmost)
```

Question3: On February 12th, 2013, which New York area airport had the windiest weather?

```python
import pandas as pd

# Load the dataset
weather = pd.read_csv("weather.csv")

# Filter for February 12, 2013
filtered = weather[
    (weather["year"] == 2013) &
    (weather["month"] == 2) &
    (weather["day"] == 12)
]

# Find airport with highest wind speed
windiest = filtered.loc[filtered["wind_speed"].idxmax()]

print(windiest[["origin", "wind_speed"]])
```

I will filter the .csv file using the wind speed which is sustained wind. Wind gust is also documented but I chose not to use this metric because wind gust is a sudden surge in wind.
